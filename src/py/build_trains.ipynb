```python
stocks = ['NSTG','ARWR','ACER','OMER','HIIQ','CLF','DDD','ICPT','AMRN','EXAS','PCRX','PTCT',
          'INSY','FATE','TNDM','DRNA','ALDX','NERV','PRQR','BOOT','FRPT','NVTA','KALV','VKTX',
          'TDOC','NVCR','RETA','TWLO','TCMD','TTD',
          'TPL', 'NVDA', 'TTD', 'AMZN', 'ALGN', 'BABA','AAPL', 'CTSH', 'ABMD', 'CRM']
```

```python
def generate_trains(col = '0.01', limit=0.01):
    # calculate lambdas
    lambs = [m.show_poisson(s) for s in stocks]
    # sort stocks by
    sorted_stocks = [(l,s) for l, s in sorted(list(zip(lambs, stocks)), reverse=True)]

    events = [m.get_events(s[1], col=col, limit=limit) for s in sorted_stocks]
    max_idx = 0
    max_len = 0
    for i, event in enumerate(events):
        if max_len < len(event):
            max_idx = i
            max_len = len(event)

    print(max_idx, sorted_stocks[max_idx])

    df = pd.DataFrame(columns=[s[1] for s in sorted_stocks])
    df['date'] = np.flip(np.array(pd.DataFrame(data=m.get_data(sorted_stocks[max_idx][1])[1])['date']), axis=0)
    df.set_index('date', inplace=True)
    df.head()

    data = {}
    max_len = 0
    for se in list(zip(sorted_stocks, events)):
        data[se[0][1]]=se[1]
        max_len = max(len(se[1]), max_len)

    print('>>>>',max_len)
    for keys in data.keys():
        #df[keys] =
        data_to_append =  np.append(data[keys], [0]*(max_len - len(data[keys])))
        #print(len(data_to_append), len(data[keys]))
        df[keys] = data_to_append


    df.to_csv('train_dates_'+col.replace('.','_')+'.csv')
```
